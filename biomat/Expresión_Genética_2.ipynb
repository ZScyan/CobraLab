```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
betas = [0.5, 1.0, 2.0]  # diferentes tasas máximas de transcripción
Kd = 10    # constante de disociación
R = np.linspace(0, 50, 500)  # rango de concentraciones de represor

# Graficar con diferentes betas
plt.figure(figsize=(7,5))
for beta in betas:
    f_R = beta / (1 + R/Kd)
    plt.plot(R, f_R, linewidth=2, label=fr"$\beta={beta}$")

plt.title("Tasa de transcripción f([R]) con diferentes β", fontsize=14)
plt.xlabel("Concentración de represor [R]")
plt.ylabel("f([R])")
plt.grid(True, linestyle="--", alpha=0.6)
plt.axvline(Kd, color="red", linestyle="--", alpha=0.7, label=r"$K_d$")
plt.legend()
plt.show()
```
